### Code for random plots used in the report 

```python
import pandas as pd
import re
import matplotlib.pyplot as plt
import numpy as np
import csv
import random
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.animation as animation
import seaborn as sns
import matplotlib.gridspec as gridspec


from BSE import market_session
```

```python
# Define the depth levels and prices for bid and ask sides
bid_prices = [-1, -2, -3, -4, -6, -8]  # Prices for bids
bid_depths = [2, 4, 3, 2, 1, 1]  # Depths for bids

ask_prices = [1, 2, 3, 5, 6, 7]  # Prices for asks
ask_depths = [1, 2, 3, 2, 3, 4]  # Depths for asks

# Convert to arrays for plotting
bid_prices = np.array(bid_prices)
ask_prices = np.array(ask_prices)
bid_depths = np.array(bid_depths)
ask_depths = np.array(ask_depths)

# Create the figure and axis
fig, ax = plt.subplots(figsize=(20, 6))

# Plot bid side (negative x-axis)
ax.bar(bid_prices, -bid_depths, color="blue", edgecolor="black", label="Buy Limit Order")

# Plot ask side (positive x-axis)
ax.bar(ask_prices, ask_depths, color="red", edgecolor="black", label="Sell Limit Order")

# Highlight mid-price, bid-price, and ask-price
mid_price = 0
ax.axhline(y=0, color="black", linestyle="-", linewidth=1)

# Add axis labels
#ax.set_xlabel("Price (USD)")
ax.set_ylabel("Depth", fontsize = 16)
ax.set_xticks([])
ax.set_xticklabels([])

# Set axis limits
ax.set_xlim(-10, 10)
ax.set_ylim(-10, 10)

ax.set

# Add legend
ax.legend(fontsize = 14)

ax.tick_params(axis='both', which='major', direction='in', length=10, width=1.5, labelsize=14,
                    top=True, bottom=True, left=True, right=False)



# Apply thicker border edges to both subplots
for ax in fig.get_axes():
    for spine in ax.spines.values():
        spine.set_linewidth(1.5)  # Adjust thickness (2 is thicker, increase if needed)

# Show the plot
plt.show()
```
